# Importing libraries and reading the file

```python
#Importing the libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib

plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #Adjusts the configuration of the plot

#Reading the dataset

df = pd.read_csv(r'C:\Users\User\Desktop\Projects\Portfolio\movies.csv')
```

```python
#Looking the first 5 rows

df.head()
```

# Cleaning and Checking the dataset

```python
#Checking for missing or null values

df.isnull().sum()
```

Note : There are no null values

```python
#Checking the datatype of the columns

print(df.dtypes)
```

```python
#Converting floating datatype to integer for ease of analysis

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df.head()
```

```python
# Ordering the data with respect to gross revenue

df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
#Displaying the whole dataset

pd.set_option('display.max_rows', None)
df.head()
```

```python
#Droping any duplicate values

df.drop_duplicates()
```

**Insight : There are no duplicate values in the dataset**

# Analysis

```python
#Using regplot to find the correlation between gross and budget

sns.regplot(x="budget", y="gross", data=df, scatter_kws={'color':'red'}, line_kws={'color':'black'}).set(title="Budget vs Gross")
```

**Insight : There is a strong positive correlation between budget and gross revenue**

```python
#Using regplot to find the correlation between score and gross

sns.regplot(x="score", y="gross", data=df, scatter_kws={'color':'red'}, line_kws={'color':'black'}).set(title="Score vs Gross")
```

**Insight : There is a moderate positive correlation between score and gross revenue**

```python
# Correlation Matrix between all numeric columns in the dataset

df.corr(method ='pearson')
```

```python
#Using a heatmap to see the correlation matrix between all numeric columns in the dataset

correlation_matrix = df.corr(method = 'pearson')
sns.heatmap(correlation_matrix, annot = True)
plt.title("Correlation Matrix For Numeric Features")
plt.xlabel("Movie features")
plt.ylabel("Movie features")
plt.show()
```

```python
#Numerizing the company column

df_numerized = df


for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
#Unstacking the df_numerized for analysis

correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
sorted_pair = corr_pairs.sort_values()
sorted_pair.head()
```

### Insight : Votes and budget have the highest correlation to gross earnings
